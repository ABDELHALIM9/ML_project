## work with outlier 

```python
def outliers(df,ft):
    q1 = df[ft].quantile(0.25)
    q3 = df[ft].quantile(0.75)
    iqr = q3 - q1 
    
    lower_limit = q1 - iqr *1.5
    upper_limit = q3 + iqr *1.5
    
    ls = df.index[(df[ft]<lower_limit) | (df[ft]>upper_limit)]
    
    return ls
```

```python
index_list = []
index_list.extend(outliers(df,'duration'))
```

```python
def remove(df,ls):
    ls = sorted(set(ls))
    df = df.drop(ls)
    return df
```

```python
df = remove(df,index_list)
```

## Prepare the data

```python
# Encode  the data 
column=['airline','source_city','departure_time','stops','arrival_time','destination_city','class']

df[column] = df[column].apply(LabelEncoder().fit_transform)

```

```python
# x and y
x=df.drop(['flight','price'],axis=1)
y=df['price']
```

```python
# split data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
print('x_train size',x_train.shape)
print('y_train size',y_train.shape)
print('x_test size',x_test.shape)
print('y_test size',y_test.shape)
```

## Model

### LinearRegression

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```
